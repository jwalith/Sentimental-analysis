```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


dataset = pd.read_csv('project1.csv')
```

```python
dataset = dataset.dropna()
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
data=[]
```

```python
for i in range(0,514):
        text=dataset['text'][i]
        text = re.sub('[^a-zA-Z]', ' ', text)
        text = text.lower()
        text = text.split()
        text = [ps.stem(word) for word in text if not word in set(stopwords.words('english'))]
        text = ' '.join(text)
        data.append(text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500)
X = cv.fit_transform(data).toarray()
```

```python
y = dataset.iloc[:,0].values
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb1=LabelEncoder()

y =lb1.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
X_test.shape
```

```python
y.shape
```

```python

from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
```

```python
model.add(Dense(output_dim = 1000, init = 'uniform', activation = 'relu', input_dim = 1487))

model.add(Dense(output_dim = 1000, init = 'uniform', activation = 'relu'))

model.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])


```

```python
model.fit(X_train,y_train,batch_size=200,epochs=1000)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred =y_pred>0.5
```
